```python
from bs4 import BeautifulSoup
# except ImportError:
#     from bs4 import BeautifulSoup


file = 'html'
h = open(file,'r')
html = h.read()
h.close()

print(html)
```

```python
parsed_html = BeautifulSoup(html, "html.parser")

print(parsed_html)
```

```python
parsed_school = parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900') ##achar escolas
school = []
for s in parsed_school:
    #s = 
    #print(s.get_text())
    school.append(s.get_text())
    
print(school)    
    
#print(parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900'))
#print(parsed_html.find('h3',class_='font-bold text-xl mb-2 text-teal-900'))
```

```python
parsed_schoolType = parsed_html.find_all('p',class_='text-base font-bold text-gray-900 mb-6') ##tipo de escola (eg, estadual)
schoolType = []
for s in parsed_schoolType:
    #s = 
    #print(s.get_text())
    schoolType.append(s.get_text())
    
print(schoolType)    
    
#print(parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900'))
#print(parsed_html.find('h3',class_='font-bold text-xl mb-2 text-teal-900'))
```

```python
parsed_ende = parsed_html.find_all('span',class_='block text-gray-900 text-sm') ##endereço
ende = []

for s in parsed_ende:
    ende.append(s.get_text())
    
print(ende)    
    
#print(parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900'))
#print(parsed_html.find('h3',class_='font-bold text-xl mb-2 text-teal-900'))
```

```python
parsed_bairro = parsed_html.find_all('span',class_='block text-orange-600 text-sm') ##bairro
bairro = []

for s in parsed_bairro:
    bairro.append(s.get_text())
    
print(bairro)    
    
#print(parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900'))
#print(parsed_html.find('h3',class_='font-bold text-xl mb-2 text-teal-900'))
```

```python
parsed_link = parsed_html.find_all('a',class_='') ##link
links = []

for s in parsed_link:
    links.append(s.get('href'))
    
print(links)    
    
#print(parsed_html.find_all('h3',class_='font-bold text-xl mb-2 text-teal-900'))
#print(parsed_html.find('h3',class_='font-bold text-xl mb-2 text-teal-900'))
```

```python
import pandas as pd

columns = ['Escola','SchoolType','Endereço_full','Rua','Numero','Complemento','Bairro_full','Bairro',
           'Cidade','Estado','Link_original']

df = pd.DataFrame(columns=columns,data={
    'Escola': school,
    'SchoolType': schoolType,
    'Endereço_full': ende,
    'Bairro_full': bairro,
    'Link_original': links
    })

df['Link_original'] = 'https://www.escol.as' + df['Link_original']

df.head(5)
```

```python
end_split = df['Endereço_full'].str.split(',',expand=True)
end_split = end_split.rename(columns={0: 'Rua',1:'Numero',2:'Complemento'})

bairro_split = df['Bairro_full'].str.split(',',expand=True)
bairro_split[['Cidade','Estado']] = bairro_split.iloc[:,1].str.split('-',expand=True)
bairro_split = bairro_split[[0,'Cidade','Estado']]
bairro_split = bairro_split.rename(columns={0:'Bairro'})


end_split.head()
```

```python
df[['Rua','Numero','Complemento']] = end_split
df[['Bairro','Cidade','Estado']] = bairro_split
```

```python
df_clean = df.drop(columns=['Endereço_full','Bairro_full'])
```
